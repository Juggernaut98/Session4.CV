```python
import numpy
from math import pi, e
import matplotlib.pyplot
```

```python
def GaussianFilter(size, sigma):
    Kernel =  numpy.fromfunction(lambda x, y: (1/(2*pi*sigma**2)) * e ** ((-1*((x-(size-1)/2)**2+(y-(size-1)/2)**2))/(2*sigma**2)), (size, size))
    return Kernel / numpy.sum(Kernel)
```

```python
GaussianFilter(4, 2)
```

```python
def GaussianBlur(img, iSize, iSigma):  
    kernel = GaussianFilter(iSize, iSigma)
    e = int((iSize-1)/2)
    Gaussian = numpy.zeros(img.shape[0]-2*e, img.shape[1]-2*e)
    for y in range(e, img.shape[0]-e):
        for x in range(e, img.shape[1]-e):
            gaussian[y-e][x-e] = numpy.sum(numpy.multiply(img[y-e:y+e+1, x-e:x+e+1], kernel))
    return Gaussian
```

```python
img = numpy.random.rand(100,100)
```

```python
# Resulting image

matplotlib.pyplot.imshow(img)
```

```python
blur_img = gaussianBlur(img, 5, 3)
```
